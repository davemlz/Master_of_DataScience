Para esta tarea crearás un script que predice los precios de casas usando kernel ridge regression. Los datos para esta tarea se encuentran en https://www.openml.org/d/574

Carga los datos usando la función "fetch_openml".

```python
from sklearn.datasets import fetch_openml

data = fetch_openml("house_16H")
```

Preprocesa los datos y realiza alguna visualización.

```python
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

scaledData = StandardScaler().fit_transform(data['data'])

print("Scatter of each variable against the price")
fig= plt.figure(figsize=(15,5))
for i in range(16):
    plt.subplot(2,8,i + 1)
    plt.subplots_adjust(hspace = 0.5)
    plt.scatter(scaledData[:,i],data['target'])
    plt.xlabel(data['feature_names'][i])    
```

Separa los datos en train (2/3) y test (1/3).

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(scaledData,data['target'],test_size = 1/3)
```

Entrena un regresor lineal para predecir la variable target (el precio) a partir de las 16 características de entrada. Calcula el Mean Absolute Error (MAE) sobre el conjunto de test. Debería ser alrededor de 25000 EUR.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae

reg = LinearRegression().fit(X_train,y_train)

y_pred = reg.predict(X_test)

print("Mean Absolute Error:",mae(y_test,y_pred))
```

Ahora entrena el método kernel ridge regression usando la aproximación Nyström o RFF y reporta su MAE sobre el conjunto de test. Ojo: el regresor en este caso consiste en un Pipeline que contiene la aproximación seguida por una regresión Ridge (lineal); No contiene el método KernelRidge explícitamente. Para encontrar parámetros adecuados puedes usar el método GridSearchCV.

```python
from sklearn.kernel_approximation import Nystroem
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV

nystroem_approx_krr = Pipeline([('feature_map',Nystroem()),('ridge',LinearRegression())])
param_grid = {'feature_map__gamma': [0.001,0.01,0.1,1,100,1000],'feature_map__n_components': [10,25,50,75,100]}
grid = GridSearchCV(nystroem_approx_krr,param_grid,cv = 5)
grid.fit(X_train,y_train)
regressor = grid.best_estimator_
y_pred = regressor.predict(X_test)

print("Mejores estimadores:",grid.best_params_)
print("Mean Absolute Error:",mae(y_test,y_pred))
```
