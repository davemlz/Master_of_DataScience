```python
import numpy as np

def reweight_distribution(original_distribution, temperature=0.5):   
    distribution = np.log(original_distribution) / temperature
    distribution = np.exp(distribution)
    return distribution / np.sum(distribution)                        
```

```python
import matplotlib.pyplot as plt

a=[1,2,3,4,5]
p_orig=[0.1, 0.1, 0.1, 0.2, 0.5]


plt.bar(a, p_orig)
plt.title("Original probability distribution")
plt.show()


p=reweight_distribution(p_orig, 0.01)

plt.bar(a, p)
plt.title("Temperature 0.01: low entropy")
plt.show()

p=reweight_distribution(p_orig, 0.3)

plt.bar(a, p)
plt.title("Temperature 0.3: intermediate entropy")
plt.show()

p=reweight_distribution(p_orig, 0.8)

plt.bar(a, p)
plt.title("Temperature 0.8: intermediate entropy")
plt.show()


p=reweight_distribution(p_orig, 4)

plt.bar(a, p)
plt.title("Temperature 4: high entropy")
plt.show()
```
