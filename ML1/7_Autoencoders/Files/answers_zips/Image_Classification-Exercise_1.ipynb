```python
import numpy as np
import matplotlib.pylab as plt
import keras
from keras.datasets import mnist, cifar10
```

**Download the datasets**

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_mnist, y_mnist = x_train, y_train

(x_train, y_train), (x_test, y_test) = cifar10.load_data()
x_cifar, y_cifar = x_train, y_train
```

## Exercice 1.1 Visualize MNIST

```python
x, y = x_mnist, y_mnist
classes = list(set(y))
print('X shape: {}; y shape: {}'.format(x.shape, y.shape))

plot_side = int(np.sqrt(len(classes))) + 1
fig = plt.figure(figsize = (15, 12))

# Plot classes samples
for i, cla in enumerate(classes):
  ind = np.argwhere(y==i)[0]
  sample = np.squeeze(x[ind, :, :])
  
  ax = fig.add_subplot(plot_side, plot_side, i+1)
  ax.grid('off')
  ax.imshow(sample)
  ax.set_title('Class {}'.format(cla))

plt.tight_layout()
plt.show()
```
